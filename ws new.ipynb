# WEB SCRAPING

# importing libraries

```python
import requests
from bs4 import BeautifulSoup
```

# scrabing a website

```python
url="https://www.flipkart.com/refrigerators/pr?sid=j9e,abm,hzg"
```

```python
url
```

# website response

```python
res=requests.get(url)
```

```python
res
```

# parsing

```python
soup=BeautifulSoup(res.text)
```

```python
soup
```

# to select a particular portion

```python
products=soup.select('._2_R_DZ span , ._4rR01T')
```

```python
products
```

```python
products_title=[]
```

```python
for item in products:
    products_title.append(item.text)
```

```python
products_title
```

```python
price=soup.select('._1_WHN1')
```

```python
price
```

# product price

```python
products_price=[]
```
